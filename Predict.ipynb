```python
classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
# Plot ad hoc CIFAR10 instances
from keras.datasets import cifar10
from keras import backend as K
from matplotlib import pyplot
from scipy.misc import toimage

#load data
K.set_image_data_format('channels_first')
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# create a grid of 3x3 images
for i in range(0, 9):
    pyplot.imshow(toimage(X_train[i]))
    #show the plot
    pyplot.show()
    print 'class: ', classes[y_train[i][0]]
```

```python
from keras.models import load_model

model = load_model('larger-cnn.h5')

model.summary()
```

```python
import numpy as np

# create a grid of 3x3 images
for i in range(0, 9):
    pyplot.imshow(toimage(X_test[i]))
    #show the plot
    pyplot.show()
    
    x = X_test[i].astype('float32') 
    x = x / 255.0 
    x = np.expand_dims(x, axis=0)
     
    pred = list(model.predict(x, verbose=0)[0])
        
    print 'class:', classes[y_test[i][0]], 'prediction:', classes[pred.index(max(pred))]
```
